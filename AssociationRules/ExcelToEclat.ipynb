```python
import pandas as pd
import numpy as np
from fim import eclat
from sklearn.feature_extraction import DictVectorizer
vec = DictVectorizer()
```

```python
trans=pd.read_excel('testdata.xlsx', index_col=0)
trans
```

```python
df_sparse = pd.DataFrame(vec.fit_transform(trans.to_dict('records')).toarray(), columns=vec.get_feature_names())
df_sparse.replace(np.NaN,0, inplace=True)
for col in df_sparse.columns:
    df_sparse[col].replace(1,col, inplace=True)
df_sparse
```

```python
dataset = [None]*len(df_sparse.index)
for i, row in enumerate(df_sparse.values):
    x = np.array(row)
    dataset[i]=x[x!=0]
dataset
```

```python
eclat(dataset, zmin=2, report='as')
```
